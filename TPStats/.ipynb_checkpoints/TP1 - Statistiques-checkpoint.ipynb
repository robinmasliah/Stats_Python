### Partie 1

```python
from os import path
import pandas as pd
import urllib
import zipfile
import sys
%matplotlib inline
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00235/'
filename = 'household_power_consumption'
zipfilename = filename + '.zip'
Location = url + zipfilename
#testing existence of file:
if sys.version_info >= (3, 0):
    if not(path.isfile('zipfilename')):
        urllib.request.urlretrieve(Location, zipfilename)
    else:
        if not(path.isfile('zipfilename')):
            urllib.urlretrieve(Location, zipfilename)
#unzip part
zip = zipfile.ZipFile(zipfilename)
zip.extractall()
na_values = ['?','']
fields = ['Date','Time','Global_active_power']
df = pd.read_csv(filename + '.txt', sep = ';', nrows = 200000, na_values = na_values, usecols = fields)
```

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

### Q1 - Gestion des données manquantes après l'importation des données

```python
#Renommage des entêtes des colonnes de df
df.columns = ['Date','Time','Global_active_power']
```

```python
#Compte des valeurs manquantes (VM)
cpt = 0
for el in df['Global_active_power']:
    if (pd.isnull(el) == True):
        cpt = cpt + 1 

print(cpt)
```

```python
#Copie du data frame orginal afin de pour modifier les données sans modifier df de départ
df2 = df
```

### Q4 - Moyennes journalières entre le 1er janvier et le 30 avril 2007

### Q6 - Time Series pandas des températures journalières entre le 1er janvier et le 30 avril 2007. Affichage sur le même graphique des températures et de la puissance énergétique consommée.

```python
df6.columns = ['Date','DEG_CELSIUS']
#print(data['Date'])
```

```python
df6["Date"] = pd.to_datetime(df6["Date"], format = "%Y%m%d")
df6.set_index('Date', inplace=True)
```

```python
#data.columns = ['Date','DEG_CELSIUS']
df6.head()
```

```python
#Simple plot
df6.plot()
```

```python
#Recalibrage sur le bon interval de temps
ax = df6.plot()
ax.set_xlim(pd.Timestamp('2007-01-01'), pd.Timestamp('2007-04-30'))
```

```python
#Rappel : dataframe contenant global_active_power
df3.head()
```

```python
#Rappel : dataframe contenant deg_celsius
df6.head()
```

```python
#Réduction des données selon le bon interval de temps
#Je crée une variable de début et une variables de fin pour les deux dataframes 


start = df3.index.searchsorted(pd.Timestamp('2007-01-01'))
end = df3.index.searchsorted(pd.Timestamp('2007-04-30'))

start2 = df6.index.searchsorted(pd.Timestamp('2007-01-01'))
end2 = df6.index.searchsorted(pd.Timestamp('2007-04-30'))

daily_GAP = df3.ix[start:end]
daily_deg_C = df6.ix[start2:end2]

daily_GAP.head()
```

```python
daily_deg_C.head()
```

```python
dft = pd.concat([daily_deg_C, daily_GAP], axis = 1)
```

```python
# Affichage du graphique des Moyennes journalières de la Consommations Electriques et de la Température


# Selection des données à afficher
conso = daily_GAP
temp = daily_deg_C  # Pas besoin de 'resample' ici puisque les données sont déjà fournies par jour

# Paramètres du graphique
fig = plt.figure(figsize=(18.5, 10.5))
plt.figure()

ax1 = fig.add_subplot(111) # add a subplot to the new figure, 111 means "1x1 grid, first subplot"
ax1.set_title('Consommations Electriques et Températures Journalières', fontsize = 20)
ax1.plot(conso, label='Global_active_power')
ax1.set_ylabel('Puissance active', fontsize=14)
ax1.legend(loc=2)

ax2 = ax1.twinx()
ax2.plot(temp, 'r-', label='DEG_CELSIUS')
ax2.set_ylabel('Température en (°C)', fontsize=14, color='r')
ax2.legend()
ax2.grid(None)
for tl in ax2.get_yticklabels():
    tl.set_color('r')

plt.show()
#dft.plot()
```

### Q7 - Visualisation de la pollution moyenne par année (pour l'ozone)

```python
df8 = pd.read_csv('20080421_20160927-PA13_auto.csv', skiprows = 1,
                  sep = ';', na_values = ['n/d'],
                  parse_dates = True)
```

```python
df8.columns = ['date','heure','NO2', 'O3']
df8["date"] = pd.to_datetime(df8["date"], format = "%d/%m/%Y")
df8.set_index('date', inplace=True)
```

```python
df8.head()
```

```python
df8.shape
df8.count()
df8.drop('heure', axis=1, inplace=True)
```

```python
cpt = 0
for el in df8:
    if (pd.isnull(el) == True):
        cpt = cpt + 1 

print(cpt)

df8 = df8.fillna(df8.mean())
```

```python
#Plot de l'ozone
df8['O3'].plot()
```

### Q8 - Visualisation de la pollution la plus critique par année pour l'ozone et pour le dioxyde d'azote.

```python
df9 = df8.resample('A').mean()
df9.plot()
```

* Visiblement noue constatons que 2016 est une année particulèrement polluée à l'ozone.
* En revanche le dioxyde d'azote a fortement baissé cette même année mais était en 2009 au plus haut avec également un fort taux de pollution à l'ozone.
* 2009 a été une année marqué par une forte pollution des deux agents.

### Q9 - Représentation par mois de la pollution à l'ozone et au dioxyde d'azote. Quel est le mois le plus pollué (pour chacun des polluants) ?

```python
#Vision par an peu pertinente
df10 = df8.resample('m').mean()
df10.plot()
```

```python
#Vision par mois sur année 2009
start3 = df8.index.searchsorted(pd.Timestamp('2009-01-01'))
end3 = df8.index.searchsorted(pd.Timestamp('2009-12-30'))
visual_pollution = df8.ix[start3:end3]
visual_pollution.plot()
```

* On remarque que le milieu d'année est fortement pollué par l'ozone. Juin et juillet sont fortement pollués par ce dernier.
* Le début d'année janvier est fortement pollué par le dioxyde d'azote.
